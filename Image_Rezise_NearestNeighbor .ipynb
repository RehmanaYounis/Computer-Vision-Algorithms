# Steps to Implement Nearest Neighbor Algorithm
1. Read the source Image
2. Get width and height of the source Image
3. Get image and height of the new/rescaled image
4. Create a dummy np array with new height and width
5. Update value of new image array by following the below steps
   5.1 Divide each position x with the total new_idth to get the percentage position of that element with respect to new array
   5.2 Multiply result with the total source width
   5.3 This will give us which value of source should be placed in the new x,y position in new array
   5.4 Similar steps should be repeated for height as well
6. Update the x,y position/pixel of new array with sourceX, sourceY position of source array
7. Display the image

```python
# Function to Rezise image
def resize_nn(img, new_width, new_height):
    img_width= img.shape[0]
    img_height= img.shape[1]
    resized_img = np.zeros((new_width, new_height, img.shape[2]), int)
    for x in range(new_width):
        for y in range(new_height):
            sourceX= int((x/new_width) * img_width)
            sourceY= int((y/new_height) * img_height)
            nx=min(sourceX, img_width-1)
            ny=min(sourceY, img_height-1)
            resized_img[x][y]= img[sourceX][sourceY]    
    return resized_img
        
        
```
